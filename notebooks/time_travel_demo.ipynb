# Time Travel Query Demonstration

```python
#Imports
from deltalake import DeltaTable
import json
import pandas as pd
```

```python
#Connection Credentials
STORAGE_OPTIONS = {
    "endpoint_url": "http://localhost:9000",
    "aws_access_key_id": "minioadmin",
    "aws_secret_access_key": "minioadmin",
    "aws_region": "us-east-1",
    "allow_http": "true",
}
```

## Taxi Demonstration

### Version History

We first get the available layers for time travel queries, gold has seperate folders so we break down each folder into one

```python
layers = {
    "Bronze": "s3://bronze/taxi/",
    "Silver": "s3://silver/taxi/",
    
    #Gold Layer Folders
    "Gold_Borough_Summary":   "s3://gold/taxi/borough_summary/",
    "Gold_Daily_Summary":   "s3://gold/taxi/daily_summary/",
    "Gold_Hourly_Summary":   "s3://gold/taxi/hourly_summary/",
    "Gold_Individual_Day_Summary":   "s3://gold/taxi/individual_day_summary/",
    "Gold_Payment_Summary":   "s3://gold/taxi/payment_summary/",
    "Gold_Zone_Performance":   "s3://gold/taxi/zone_performance/",
}
```

For each layer we get the version history available from the logs (Last 5)

```python
print("Getting version histories for each layer")

for name, uri in layers.items():
    dt = DeltaTable(uri, storage_options=STORAGE_OPTIONS)
    history = dt.history(5)
    
    print(f"\n=== {name} — {len(dt.history())} versions ===")
    df = pd.DataFrame(history)[["version", "timestamp", "operation"]]

    display(df)

    print("Details of Last Operation: ")
    print(json.dumps(history[0], indent=2, default=str))
```

#### Bronze Time Travel

We check what the bronze table looked like after the first 5 tables were loaded.

A predicate indicates the file for that predicate was overwritten at that version.

```python
dt = DeltaTable(layers["Bronze"], storage_options=STORAGE_OPTIONS)
history = dt.history()

version_to_info = {
    entry["version"]: {
        "timestamp": pd.to_datetime(entry["timestamp"], unit='ms'),
        "predicate": entry.get("operationParameters", {}).get("predicate", "N/A")
    }
    for entry in history
}

for version in range(0, 5):
    dt_v = DeltaTable(
        layers["Bronze"],
        version=version,
        storage_options=STORAGE_OPTIONS
    )
    count = dt_v.to_pyarrow_dataset().count_rows()
    info = version_to_info[version]
    ts = info["timestamp"].strftime("%B %d, %Y %H:%M:%S")
    predicate = info["predicate"]
    schema_fields = ", ".join(
        f"{field.name}: {field.type}" 
        for field in dt_v.schema().fields
    )

    print(f"{'─' * 60}")
    print(f"  Version   : {version}")
    print(f"  Ingested  : {ts}")
    print(f"  Predicate : {predicate}")
    print(f"  Row count : {count:,}")
    print(f"  Schema    : {schema_fields}")

print(f"{'─' * 60}")
```

In production, Ingested At will show the actual month and year ingested since the pipeline will only run once per month

### Silver Time Travel

Check what silver looks like after each transform is done

```python
dt = DeltaTable(layers["Silver"], storage_options=STORAGE_OPTIONS)
history = dt.history()

version_to_info = {
    entry["version"]: {
        "timestamp": pd.to_datetime(entry["timestamp"], unit='ms'),
        "predicate": entry.get("operationParameters", {}).get("predicate", "N/A")
    }
    for entry in history
}

for version in range(0, 5):
    dt_v = DeltaTable(
        layers["Silver"],
        version=version,
        storage_options=STORAGE_OPTIONS
    )
    count = dt_v.to_pyarrow_dataset().count_rows()
    info = version_to_info[version]
    ts = info["timestamp"].strftime("%B %d, %Y %H:%M:%S")
    predicate = info["predicate"]
    schema_fields = ", ".join(
        f"{field.name}: {field.type}" 
        for field in dt_v.schema().fields
    )

    print(f"{'─' * 60}")
    print(f"  Version   : {version}")
    print(f"  Ingested  : {ts}")
    print(f"  Predicate : {predicate}")
    print(f"  Row count : {count:,}")
    print(f"  Schema    : {schema_fields}")

print(f"{'─' * 60}")
```

In production this can be done using different months and timestamps for when the transform was executed in airflow

#### Schema Evolution At Bronze Layer

We check to see which columns were removed or added over time

```python
dt = DeltaTable(layers["Bronze"], storage_options=STORAGE_OPTIONS)

# Build a schema snapshot for each of the first 5 versions
version_schemas = {}
for version in range(0, 5):
    dt_v = DeltaTable(layers["Bronze"], version=version, storage_options=STORAGE_OPTIONS)
    version_schemas[version] = {field.name: str(field.type) for field in dt_v.schema().fields}

# Use version 0 as the baseline
baseline = version_schemas[0]
baseline_cols = set(baseline.keys())

print(f"{'─' * 60}")
print(f"  Baseline schema (version 0) — {len(baseline_cols)} columns")
for name, dtype in baseline.items():
    print(f"    {name}: {dtype}")

for version in range(1, 5):
    current_cols = set(version_schemas[version].keys())
    info = version_to_info[version]
    ts = info["timestamp"].strftime("%B %d, %Y %H:%M:%S")

    added   = current_cols - baseline_cols
    removed = baseline_cols - current_cols
    changed = {
        col for col in current_cols & baseline_cols
        if version_schemas[version][col] != baseline[col]
    }

    print(f"{'─' * 60}")
    print(f"  Version {version} | {ts}")

    if not added and not removed and not changed:
        print(f"    No schema changes from version 0")
    if added:
        for col in added:
            print(f"    + ADDED   {col}: {version_schemas[version][col]}")
    if removed:
        for col in removed:
            print(f"    - REMOVED {col}: {baseline[col]}")
    if changed:
        for col in changed:
            print(f"    ~ CHANGED {col}: {baseline[col]} -> {version_schemas[version][col]}")

print(f"{'─' * 60}")
```
